## removing outliers

```python
percentile25_exp = df['Expense'].quantile(0.25)
percentile75_exp = df['Expense'].quantile(0.75)
iqr_exp=percentile75_exp-percentile25_exp

upper_limit_exp = percentile75_exp + 1.5 * iqr_exp
lower_limit_exp = percentile25_exp - 1.5 * iqr_exp

upper_exp = np.where(df['Expense'] >= upper_limit_exp)
lower_exp = np.where(df['Expense'] <=lower_limit_exp) 

#df=df.drop([upper_exp[0],inplace=True)
#df=df.drop(lower_exp[0], inplace = True)
            
df = df[df['Expense'].between(lower_limit_exp,upper_limit_exp)]
df.shape
```

```python
percentile25_s3 = df['Score3'].quantile(0.25)
percentile75_s3 = df['Score3'].quantile(0.75)
iqr_s3=percentile75_s3-percentile25_s3

upper_limit_s3 = percentile75_s3 + 1.5 * iqr_s3
lower_limit_s3 = percentile25_s3 - 1.5 * iqr_s3


upper_s3 = np.where(df['Score3'] >= upper_limit_s3)
lower_s3 = np.where(df['Score3'] <=lower_limit_s3) 

df = df[df['Score3'].between(lower_limit_s3,upper_limit_s3)]
df.shape
```

```python

percentile25_s1 = df['Score1'].quantile(0.25)
percentile75_s1 = df['Score1'].quantile(0.75)
iqr_s1=percentile75_s1-percentile25_s1

upper_limit_s1 = percentile75_s1 + 1.5 * iqr_s1
lower_limit_s1 = percentile25_s1 - 1.5 * iqr_s1

upper_s1 = np.where(df['Score1'] >= upper_limit_s1)
lower_s1 = np.where(df['Score1'] <=lower_limit_s1) 

df = df[df['Score1'].between(lower_limit_s1,upper_limit_s1)]
df.shape
```

```python


percentile25_i = df['Income'].quantile(0.25)
percentile75_i = df['Income'].quantile(0.75)
iqr_i=percentile75_i-percentile25_i

upper_limit_i = percentile75_i + 1.5 * iqr_i
lower_limit_i = percentile25_i - 1.5 * iqr_i


upper_i = np.where(df['Income'] >= upper_limit_i)
lower_i = np.where(df['Income'] <=lower_limit_i)

df = df[df['Income'].between(lower_limit_i,upper_limit_i)]
df.shape
```

```python
percentile25_s2 = df['Score2'].quantile(0.25)
percentile75_s2 = df['Score2'].quantile(0.75)
iqr_s2=percentile75_s2-percentile25_s2

upper_limit_s2 = percentile75_s2 + 1.5 * iqr_s2
lower_limit_s2 = percentile25_s2 - 1.5 * iqr_s2


df = df[df['Score2'].between(lower_limit_s2,upper_limit_s2)]
df.shape
```

```python
percentile25_s4 = df['Score4'].quantile(0.25)
percentile75_s4 = df['Score4'].quantile(0.75)
iqr_s4=percentile75_s4-percentile25_s4

upper_limit_s4 = percentile75_s4 + 1.5 * iqr_s4
lower_limit_s4 = percentile25_s4 - 1.5 * iqr_s4


df = df[df['Score4'].between(lower_limit_s4,upper_limit_s4)]
df.shape
```

```python
percentile25_s5 = df['Score5'].quantile(0.25)
percentile75_s5 = df['Score5'].quantile(0.75)
iqr_s5=percentile75_s5-percentile25_s5

upper_limit_s5 = percentile75_s5 + 1.5 * iqr_s5
lower_limit_s5 = percentile25_s5 - 1.5 * iqr_s5


df = df[df['Score5'].between(lower_limit_s5,upper_limit_s5)]
df.shape
```
